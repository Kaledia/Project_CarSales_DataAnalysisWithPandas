```python

# Project to analyze data with Pandas Python: data cleaning (find and fill null, remove unwanted data), data exploration (value counts, filtering), data manipulation (applying functions), data visualisation (plot, histogram)
# Thanks to the Youtube Channel "Data Science Lovers" for the data: https://drive.google.com/file/d/1MlrQ... 

# Import operating system, numpy and pandas libraries
import os
import pandas as pd 
import numpy as np

# Check current working library
os.getcwd()

# Import csv data from local repository 
car=pd.read_csv ("Cars_Data.csv")

# Check if data was successfully imported
car
```

```python
print (car[car['Cylinders'].isnull()])
```

```python
car_noNaN= car[car.Make.notnull()]
```

```python
car_noNaN.isna().sum().plot(kind='bar')
```

```python
# As the two cars with missing cylinders values are Mazda, print Mazda data
print (car_noNaN[car_noNaN['Make'] == 'Mazda'])
```

```python
# As the 2 missing cylinders values are for a 1.3 engineSize, I assume that it is 4 cylinders (Google confirm that a Mazda RX8 is a 4 cylinder car)
car_noNaN['Cylinders'].fillna(4, inplace=True)
```
